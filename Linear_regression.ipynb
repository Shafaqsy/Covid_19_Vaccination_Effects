```python
# Load the data
file_path = Path('masterfile.csv')
df = pd.read_csv(file_path)
df.head()
```

# Linear Regression

```python
plt.scatter(df.total_doses, df.cases)
plt.xlabel('Total number of doses')
plt.ylabel('Cases in California')
plt.show()
```

```python
plt.scatter(df.cumulative_total_doses, df.cases)
plt.xlabel('Total cumulative number of doses')
plt.ylabel('Cases in California')
plt.show()
```

```python
plt.scatter(df.cumulative_total_doses, df.deaths)
plt.xlabel('Total cumulative number of doses')
plt.ylabel('Deaths in California')
plt.show()
```

```python
plt.scatter(df.total_doses, df.deaths)
plt.xlabel('Total number of doses')
plt.ylabel('Deaths in California')
plt.show()
```

```python
X = df.cumulative_total_doses.values.reshape(-1, 1)
```

```python
X[:5]
```

```python
X.shape
```

```python
y = df.deaths
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
y_pred = model.predict(X)
print(y_pred.shape)
```

```python
plt.scatter(X, y)
plt.plot(X, y_pred, color='red')
plt.show()
```

```python
print(model.coef_)
print(model.intercept_)
```
